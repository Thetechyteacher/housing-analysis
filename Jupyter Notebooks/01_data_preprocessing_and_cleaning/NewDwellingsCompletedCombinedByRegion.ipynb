## Combining data for New Dwellings Completed by Inner, Outer, City of London Regions and saving data for all years (2006-2022) together in one dataset:

```python
import pandas as pd
```

```python
years_2006_2007 = pd.read_csv('../../Datasets/02_clean_data_files/new_dwellings_cleaned_data/new_dwellings_completed_pre_FinCrisis2008_period.csv')
years_2006_2007
```

```python
years_2008_2009 = pd.read_csv('../../Datasets/02_clean_data_files/new_dwellings_cleaned_data/new_dwellings_completed_FinCrisis2008_period.csv')
years_2008_2009
```

```python
years_2010_2012 = pd.read_csv('../../Datasets/02_clean_data_files/new_dwellings_cleaned_data/new_dwellings_completed_2010-2012_period.csv')
years_2010_2012
```

```python
years_2013_2016 = pd.read_csv('../../Datasets/02_clean_data_files/new_dwellings_cleaned_data/new_dwellings_completed_2013-2016_period.csv')
years_2013_2016
```

```python
years_2017_2019 = pd.read_csv('../../Datasets/02_clean_data_files/new_dwellings_cleaned_data/new_dwellings_completed_pre_covid19_period.csv')
years_2017_2019
```

```python
years_2020_2022 = pd.read_csv('../../Datasets/02_clean_data_files/new_dwellings_cleaned_data/new_dwellings_completed_covid19_period.csv')
years_2020_2022
```

```python
new_dwellings_2006_2022 = pd.concat([years_2006_2007, years_2008_2009, years_2010_2012, years_2013_2016, years_2017_2019, years_2020_2022], ignore_index=True,axis=0)
new_dwellings_2006_2022 
```

```python
new_dwellings_2006_2022.drop('Unnamed: 0', axis=1, inplace=True)
new_dwellings_2006_2022
```

### Adding a new column to the dataframe that will represent the right region for each district in the dataframe. 
Inner London: Camden, Greenwich, Hackney, Hammersmith and Fulham, Islington, Kensington and Chelsea, Lambeth, Lewisham, Southwark, Tower Hamlets, Wandsworth, Westminster.
<br/>City of London: City of London.
<br/>Outer London: Barking and Dagenham, Barnet, Bexley, Brent, Bromley, Croydon, Ealing, Enfield, Haringey, Harrow, Havering, Hillingdon, Hounslow, Kingston upon Thames, Merton, Newham, Redbridge, Richmond upon Thames, Sutton, Waltham Forest.tminster


```python
district_categories = {
    'Camden': 'Inner London', 'Greenwich': 'Inner London', 'Hackney': 'Inner London', 'Hammersmith and Fulham': 'Inner London', 'Islington': 'Inner London', 
    'Kensington and Chelsea': 'Inner London', 'Lambeth': 'Inner London', 'Lewisham': 'Inner London', 'Southwark': 'Inner London', 'Tower Hamlets': 'Inner London',
    'Wandsworth': 'Inner London', 'Westminster': 'Inner London', 'City of London': 'City of London', 
    'Barking and Dagenham': 'Outer London', 'Barnet': 'Outer London', 'Bexley': 'Outer London', 'Brent': 'Outer London', 'Bromley': 'Outer London',
    'Croydon': 'Outer London', 'Ealing': 'Outer London', 'Enfield': 'Outer London', 'Haringey': 'Outer London', 'Harrow': 'Outer London',
    'Havering': 'Outer London', 'Hillingdon': 'Outer London', 'Hounslow': 'Outer London', 'Kingston upon Thames': 'Outer London', 'Merton': 'Outer London',
    'Newham': 'Outer London', 'Redbridge': 'Outer London', 'Richmond upon Thames': 'Outer London', 'Sutton': 'Outer London', 'Waltham Forest': 'Outer London'
}

# Add a new column for district categories
new_dwellings_2006_2022['Region'] = new_dwellings_2006_2022['Local Authority Districts'].map(district_categories)

# display updated DataFrame
new_dwellings_2006_2022.head(33)
```

```python
# changing the region column to be the first one
column_region = new_dwellings_2006_2022.pop("Region")
new_dwellings_2006_2022.insert(0, "Region", column_region)
new_dwellings_2006_2022
```

```python
#checking if any values are null
new_dwellings_2006_2022.isnull().values.any()
```

```python
new_dwellings_2006_2022.drop(columns=['Local Authority Districts'], inplace=True)
```

```python
new_dwellings_2006_2022
```

Combining data by Inner London, Outter London.

```python
new_dwellings_completed_by_region= new_dwellings_2006_2022.groupby(['Region', 'Year']).sum().reset_index()
new_dwellings_completed_by_region
```

```python
new_dwellings_completed_by_region['All'] = new_dwellings_completed_by_region[['Private Enterprise', 'Housing Associations', 'Local Authority']].sum(axis=1)
new_dwellings_completed_by_region
```

```python
new_dwellings_completed_by_region.to_csv('../../Datasets/02_clean_data_files/new_dwellings_completed_by_region.csv')
```
