combining data for 2010

```python
common_columns = ['Private Enterprise', 'Housing Associations', 'Local Authority']
year_2010 = year_2010_Q1.copy()
year_2010[common_columns] = year_2010_Q1[common_columns].add(year_2010_Q2[common_columns]).add(year_2010_Q3[common_columns]).add(year_2010_Q4[common_columns])
#no need to use fill_value=0 in .add() as there is no missing values
year_2010.insert(0, 'Year', '2010')
year_2010
```

combining data for 2011

```python
common_columns = ['Private Enterprise', 'Housing Associations', 'Local Authority']
year_2011 = year_2011_Q1.copy()
year_2011[common_columns] = year_2011_Q1[common_columns].add(year_2011_Q2[common_columns]).add(year_2011_Q3[common_columns]).add(year_2011_Q4[common_columns])
#no need to use fill_value=0 in .add() as there is no missing values
year_2011.insert(0, 'Year', '2011')
year_2011
```

combining data for 2012

```python
common_columns = ['Private Enterprise', 'Housing Associations', 'Local Authority']
year_2012 = year_2012_Q1.copy()
year_2012[common_columns] = year_2012_Q1[common_columns].add(year_2012_Q2[common_columns]).add(year_2012_Q3[common_columns]).add(year_2012_Q4[common_columns])
#no need to use fill_value=0 in .add() as there is no missing values
year_2012.insert(0, 'Year', '2012')
year_2012
```

combining all years to get the 3 years pre-covid period

```python
years_2010_2012 = pd.concat([year_2010, year_2011, year_2012], ignore_index=True,axis=0)
years_2010_2012
```
