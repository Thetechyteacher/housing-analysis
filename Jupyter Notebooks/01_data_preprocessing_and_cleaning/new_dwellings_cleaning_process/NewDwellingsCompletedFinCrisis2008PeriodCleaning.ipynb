## Extracting relevent data for the Financial Crisis 2008 Period:

```python
import pandas as pd
year_2008_Q1 = pd.read_excel('LiveTable253a.xlsx', sheet_name='2008 Q1', skiprows=58, nrows=33, usecols="F,M:O", header=None, names=['Local Authority Districts', 'Private Enterprise', 'Housing Associations', 'Local Authority'])
year_2008_Q1
```

```python
year_2008_Q2 = pd.read_excel('LiveTable253a.xlsx', sheet_name='2008 Q2', skiprows=58, nrows=33, usecols="F,M:O", header=None, names=['Local Authority Districts', 'Private Enterprise', 'Housing Associations', 'Local Authority'])
year_2008_Q2
```

```python
year_2008_Q3 = pd.read_excel('LiveTable253a.xlsx', sheet_name='2008 Q3', skiprows=58, nrows=33, usecols="F,M:O", header=None, names=['Local Authority Districts', 'Private Enterprise', 'Housing Associations', 'Local Authority'])
year_2008_Q3
```

```python
year_2008_Q4 = pd.read_excel('LiveTable253a.xlsx', sheet_name='2008 Q4', skiprows=58, nrows=33, usecols="F,M:O", header=None, names=['Local Authority Districts', 'Private Enterprise', 'Housing Associations', 'Local Authority'])
year_2008_Q4
```

```python
year_2009_Q1 = pd.read_excel('LiveTable253a.xlsx', sheet_name='2009 Q1', skiprows=58, nrows=33, usecols="F,M:O", header=None, names=['Local Authority Districts', 'Private Enterprise', 'Housing Associations', 'Local Authority'])
year_2009_Q1
```

```python
year_2009_Q2 = pd.read_excel('LiveTable253a.xlsx', sheet_name='2009 Q2', skiprows=68, nrows=33, usecols="F,M:O", header=None, names=['Local Authority Districts', 'Private Enterprise', 'Housing Associations', 'Local Authority'])
year_2009_Q2
```

```python
year_2009_Q3 = pd.read_excel('LiveTable253a.xlsx', sheet_name='2009 Q3', skiprows=68, nrows=33, usecols="F,M:O", header=None, names=['Local Authority Districts', 'Private Enterprise', 'Housing Associations', 'Local Authority'])
year_2009_Q3
```

```python
year_2009_Q4 = pd.read_excel('LiveTable253a.xlsx', sheet_name='2009 Q4', skiprows=68, nrows=33, usecols="F,M:O", header=None, names=['Local Authority Districts', 'Private Enterprise', 'Housing Associations', 'Local Authority'])
year_2009_Q4
```

combining data for 2008

```python
common_columns = ['Private Enterprise', 'Housing Associations', 'Local Authority']
year_2008 = year_2008_Q1.copy()
year_2008[common_columns] = year_2008_Q1[common_columns].add(year_2008_Q2[common_columns]).add(year_2008_Q3[common_columns]).add(year_2008_Q4[common_columns])
#no need to use fill_value=0 in .add() as there is no missing values
year_2008.insert(0, 'Year', '2008')
year_2008
```

combining data for 2009

```python
common_columns = ['Private Enterprise', 'Housing Associations', 'Local Authority']
year_2009 = year_2009_Q1.copy()
year_2009[common_columns] = year_2009_Q1[common_columns].add(year_2009_Q2[common_columns]).add(year_2009_Q3[common_columns]).add(year_2009_Q4[common_columns])
#no need to use fill_value=0 in .add() as there is no missing values
year_2009.insert(0, 'Year', '2009')
year_2009
```

combining all years together

```python
years_2008_2009 = pd.concat([year_2008, year_2009], ignore_index=True,axis=0)
years_2008_2009 
```
