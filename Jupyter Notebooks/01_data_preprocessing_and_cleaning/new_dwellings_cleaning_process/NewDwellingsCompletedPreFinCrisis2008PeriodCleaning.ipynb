## Extracting relevent data for the pre Financial Crisis 2008 Period:

```python
import pandas as pd
year_2006_Q1 = pd.read_excel('LiveTable253a.xlsx', sheet_name='2006 Q1', skiprows=58, nrows=33, usecols="F,M:O", header=None, names=['Local Authority Districts', 'Private Enterprise', 'Housing Associations', 'Local Authority'])
year_2006_Q1
```

```python
year_2006_Q2 = pd.read_excel('LiveTable253a.xlsx', sheet_name='2006 Q2', skiprows=58, nrows=33, usecols="F,M:O", header=None, names=['Local Authority Districts', 'Private Enterprise', 'Housing Associations', 'Local Authority'])
year_2006_Q2
```

```python
year_2006_Q3 = pd.read_excel('LiveTable253a.xlsx', sheet_name='2006 Q3', skiprows=58, nrows=33, usecols="F,M:O", header=None, names=['Local Authority Districts', 'Private Enterprise', 'Housing Associations', 'Local Authority'])
year_2006_Q3
```

```python
year_2006_Q4 = pd.read_excel('LiveTable253a.xlsx', sheet_name='2006 Q4', skiprows=58, nrows=33, usecols="F,M:O", header=None, names=['Local Authority Districts', 'Private Enterprise', 'Housing Associations', 'Local Authority'])
year_2006_Q4
```

```python
year_2007_Q1 = pd.read_excel('LiveTable253a.xlsx', sheet_name='2007 Q1', skiprows=58, nrows=33, usecols="F,M:O", header=None, names=['Local Authority Districts', 'Private Enterprise', 'Housing Associations', 'Local Authority'])
year_2007_Q1
```

```python
year_2007_Q2 = pd.read_excel('LiveTable253a.xlsx', sheet_name='2007 Q2', skiprows=58, nrows=33, usecols="F,M:O", header=None, names=['Local Authority Districts', 'Private Enterprise', 'Housing Associations', 'Local Authority'])
year_2007_Q2
```

```python
year_2007_Q3 = pd.read_excel('LiveTable253a.xlsx', sheet_name='2007 Q3', skiprows=58, nrows=33, usecols="F,M:O", header=None, names=['Local Authority Districts', 'Private Enterprise', 'Housing Associations', 'Local Authority'])
year_2007_Q3
```

```python
year_2007_Q4 = pd.read_excel('LiveTable253a.xlsx', sheet_name='2007 Q4', skiprows=58, nrows=33, usecols="F,M:O", header=None, names=['Local Authority Districts', 'Private Enterprise', 'Housing Associations', 'Local Authority'])
year_2007_Q4
```

combining data for 2006

```python
common_columns = ['Private Enterprise', 'Housing Associations', 'Local Authority']
year_2006 = year_2006_Q1.copy()
year_2006[common_columns] = year_2006_Q1[common_columns].add(year_2006_Q2[common_columns]).add(year_2006_Q3[common_columns]).add(year_2006_Q4[common_columns])
#no need to use fill_value=0 in .add() as there is no missing values
year_2006.insert(0, 'Year', '2006')
year_2006
```

combining data for 2007

```python
common_columns = ['Private Enterprise', 'Housing Associations', 'Local Authority']
year_2007 = year_2007_Q1.copy()
year_2007[common_columns] = year_2007_Q1[common_columns].add(year_2007_Q2[common_columns]).add(year_2007_Q3[common_columns]).add(year_2007_Q4[common_columns])
#no need to use fill_value=0 in .add() as there is no missing values
year_2007.insert(0, 'Year', '2007')
year_2007
```

combining all years together:

```python
years_2006_2007 = pd.concat([year_2006, year_2007], ignore_index=True,axis=0)
years_2006_2007 
```
