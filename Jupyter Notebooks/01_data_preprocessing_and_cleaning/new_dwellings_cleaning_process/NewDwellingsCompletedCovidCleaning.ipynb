Next step is to combine data for each year

combining data for year 2020

```python
common_columns = ['Private Enterprise', 'Housing Associations', 'Local Authority']
year_2020 = year_2020_Q1.copy()
year_2020[common_columns] = year_2020_Q1[common_columns].add(year_2020_Q2[common_columns]).add(year_2020_Q3[common_columns]).add(year_2020_Q4[common_columns])
#no need to use fill_value=0 in .add() as there is no missing values
year_2020.insert(0, 'Year', '2020')
year_2020
```

combining data for year 2021

```python
common_columns = ['Private Enterprise', 'Housing Associations', 'Local Authority']
year_2021 = year_2021_Q1.copy()
year_2021[common_columns] = year_2021_Q1[common_columns].add(year_2021_Q2[common_columns]).add(year_2021_Q3[common_columns]).add(year_2021_Q4[common_columns])
year_2021.insert(0, 'Year', '2021')
year_2021
```

combining data for year 2022

```python
common_columns = ['Private Enterprise', 'Housing Associations', 'Local Authority']
year_2022 = year_2022_Q1.copy()
year_2022[common_columns] = year_2022_Q1[common_columns].add(year_2022_Q2[common_columns]).add(year_2022_Q3[common_columns]).add(year_2022_Q4[common_columns])
year_2022.insert(0, 'Year', '2022')
year_2022
```

combining all years together

```python
years_2020_2022 = pd.concat([year_2020, year_2021, year_2022], ignore_index=True,axis=0)
years_2020_2022
```
