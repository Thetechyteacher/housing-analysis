```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('../../Datasets/01_raw_data_files/CPIannualrate.xls')
```

```python
column_rename = {
    'Title': 'Year',
    'CPIH ANNUAL RATE 00: ALL ITEMS 2015=100': 'CPIH ANNUAL RATE (%)',
}

# Rename the columns
df.rename(columns=column_rename, inplace=True)
```

```python
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')  # Convert to numeric, allowing NaN
df['Year'] = df['Year'].fillna(0).astype(int)  # Fill NaN with 0 and then convert to int

```

```python
df = df[(df['Year'] >= 2006) & (df['Year'] <= 2022)]
```

```python
# Filter the data for years 2005-2022

# Plotting using Matplotlib
plt.figure(figsize=(8, 6))
plt.plot(df['Year'], df['CPIH ANNUAL RATE (%)'], marker='o', linestyle='-', color='b')
plt.title('Annual Inflation Rates (2006-2022)')
plt.xlabel('Year')
plt.ylabel('Inflation Rate (%)')
plt.grid(True)

plt.xticks(range(2006, 2023), rotation=45)
plt.tight_layout()

plt.show()
```
