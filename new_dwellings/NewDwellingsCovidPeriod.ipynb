Next step is to combine data for each year

combining data for year 2020

combining data for year 2021

combining data for year 2022

```python
common_columns = ['Private Enterprise', 'Housing Associations', 'Local Authority']
year_2022 = year_2022_Q1.copy()
year_2022[common_columns] = year_2022_Q1[common_columns].add(year_2022_Q2[common_columns]).add(year_2022_Q3[common_columns]).add(year_2022_Q4[common_columns])
year_2022.insert(0, 'Year', '2022')
year_2022
```
