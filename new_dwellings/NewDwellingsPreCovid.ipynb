Combining data for each year:

combining data for 2017

```python
common_columns = ['Private Enterprise', 'Housing Associations', 'Local Authority']
year_2017 = year_2017_Q1.copy()
year_2017[common_columns] = year_2017_Q1[common_columns].add(year_2017_Q2[common_columns]).add(year_2017_Q3[common_columns]).add(year_2017_Q4[common_columns])
#no need to use fill_value=0 in .add() as there is no missing values
year_2017.insert(0, 'Year', '2017')
year_2017
```

combining data for 2018

```python
common_columns = ['Private Enterprise', 'Housing Associations', 'Local Authority']
year_2018 = year_2018_Q1.copy()
year_2018[common_columns] = year_2018_Q1[common_columns].add(year_2018_Q2[common_columns]).add(year_2018_Q3[common_columns]).add(year_2018_Q4[common_columns])
#no need to use fill_value=0 in .add() as there is no missing values
year_2018.insert(0, 'Year', '2018')
year_2018
```

combining data for 2019

```python
common_columns = ['Private Enterprise', 'Housing Associations', 'Local Authority']
year_2019 = year_2019_Q1.copy()
year_2019[common_columns] = year_2019_Q1[common_columns].add(year_2019_Q2[common_columns]).add(year_2019_Q3[common_columns]).add(year_2019_Q4[common_columns])
#no need to use fill_value=0 in .add() as there is no missing values
year_2019.insert(0, 'Year', '2019')
year_2019
```

combining all years to get the 3 years pre-covid period

```python
years_2017_2019 = pd.concat([year_2017, year_2018, year_2019], ignore_index=True,axis=0)
years_2017_2019
```
